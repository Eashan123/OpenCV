```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

image = cv2.imread('./Images/one.jpg')
hist = cv2.calcHist([image], [2], None, [256], [0, 256])

#Plotting a histogram
plt.hist(image.ravel(), 256, [0, 256])
plt.show()

color = ('B','G','R')

for a, col in enumerate(color):
    hist2 = cv2.calcHist([image], [a], None, [256], [0, 256])
    plt.plot(hist2, color = col)
    plt.xlim([0, 256])
plt.show()

```
